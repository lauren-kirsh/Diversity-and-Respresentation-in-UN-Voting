```python
# drop unimportant columns
colonialism_df = colonialism_df.drop(['colstart.belgium_max', 'colstart.britain_max', 'colstart.france_max',
       'colstart.germany_max', 'colstart.italy_max',
       'colstart.netherlands_max', 'colstart.portugal_max',
       'colstart.spain_max', 'colend.belgium_max', 'colend.britain_max',
       'colend.france_max', 'colend.germany_max', 'colend.italy_max',
       'colend.netherlands_max', 'colend.portugal_max', 'colend.spain_max',
       'colstart.belgium_mean', 'colstart.britain_mean',
       'colstart.france_mean', 'colstart.germany_mean', 'colstart.italy_mean',
       'colstart.netherlands_mean', 'colstart.portugal_mean',
       'colstart.spain_mean', 'colend.belgium_mean', 'colend.britain_mean',
       'colend.france_mean', 'colend.germany_mean', 'colend.italy_mean',
       'colend.netherlands_mean', 'colend.portugal_mean', 'colend.spain_mean'], axis=1)
```

```python
# rename columns
colonialism_df = colonialism_df.rename(columns={'country':'countryname'})
```

```python
colonialism_df.head()
```

```python
# merge on countryname 
colonialism_df = pd.merge(colonialism_df, codes, on=['countryname'], how='left')
```

### Create new dataset and column

```python
colonialism_df['colony?'] = colonialism_df['col.belgium'] +	colonialism_df['col.britain']	+	colonialism_df['col.france'] + colonialism_df['col.germany'] +	colonialism_df['col.italy']	+	colonialism_df['col.netherlands']	+	colonialism_df['col.portugal']	+	colonialism_df['col.spain']
```

```python
colonialism_df.columns
```

```python
colonialism_df = colonialism_df.drop(['countryname', 'col.belgium', 'col.britain', 'col.france', 'col.germany',
       'col.italy', 'col.netherlands', 'col.portugal', 'col.spain'], axis = 1)
```

```python
#  if a row in "colony?"=0, then do nothing. If a row in "colony?">0, then make "colony?"=1. 
colonialism_df.loc[colonialism_df['colony?'] > 0, 'colony?'] = 1
```

```python
colonialism_df['colony?'].unique()
```
