**1) What is the difference between single-winner and multiwinner elections?**

The number of candidates and the numbers of votes per voters.

*Single-winner:* Each voter is allowed to vote for only one candidate, and the winner of the election is the candidate who represents a plurality of voters or, in other words, received the largest number of votes.


*   Only 1 vote per voter
*   Only 1 candidate can win
*   Winner is determined by plurality (1 or 0)



*Multiwinner:* Cumulative voting (also accumulation voting, weighted voting or multi-voting) is a multiple-winner cardinal voting method intended to promote more proportional representation than winner-take-all elections. Cumulative voting is used frequently in corporate governance, where it is mandated by some (7) U.S. states.


*   2 or more votes per voter
*   2 or more candidates win
*   Winners are determined by ranking (Borda)

**4) What is the novelty (or intended contribution) of the paper?**

We, as a society, create committees to make judgements for a large group of people (i.e. politics, business, etc.) Generally, we have a question of “What makes a committee fair?” Up until this point, fairness was determined by the diversity of the committee (i.e. which group were included in the committee) because there is an underlying assumption that a voter votes for a candidate with attributes that coincide with their own (i.e. women vote for women). However, that is not necessarily the case. For example, having a female candidate on the committee is different from having a candidate on the committee who is preferred by the female voters, and who themselves may or may not be female. Therefore, we should treat candidate attributes and voter attributes as two separate entities that need to be thought about when creating a fair committee. So, you guys designed a multi-election fairness model called the DiRe Committee Model that attempst to fix this problem. Essentially, diversity and representation are two separate things/goals.

```python
# function to get unique values
def unique(list1):
 
    # initialize a null list
    unique_list = []
     
    # traverse for all elements
    for x in list1:
        # check if exists in unique_list or not
        if x not in unique_list:
            unique_list.append(x)
    # print list
    for x in unique_list:
        print(x)
```
