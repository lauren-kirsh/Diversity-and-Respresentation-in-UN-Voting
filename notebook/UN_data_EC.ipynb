```python

for each_year in year_list:
  #print(each_year) # for debugging - which year does it get stuck
  #print(UN_data_df.shape)

  # create a subset of the year
  subset_year = UN_data_df[UN_data_df['year'] == each_year]

  # create a DataFrame with the year, country, and GDP (to be merged) 
  data_to_merge = pd.DataFrame({'year': each_year,
                          'Country': subset_year.Country,
                          ('GDP per capita (current US$)'): 0})
  # get the DataFrame into the proper setup
  data_to_merge.drop_duplicates(inplace=True)
  data_to_merge.reset_index(drop=True, inplace=True)

  # adding values into the DataFrame
  for country in data_to_merge.Country:
    for gdp_index in range(len(gdp_df)):
      if (country ==  gdp_df['Country'][gdp_index]): # we just need to check country because year is set for this iteration 
        index = data_to_merge[data_to_merge['Country'] == country].index.values[0]
        data_to_merge['GDP per capita (current US$)'][index] = gdp_df[str(each_year)][gdp_index]
        # change to loc ?

  # merge 
  UN_data_df = pd.merge(UN_data_df, data_to_merge, on=['year', 'Country'], how='left')
  #print(UN_data_df.columns)
```

```python
# create a data frame with these columns resolution, year, developed yes, developed no, undeveloped yes, undeveloped no
vote_by_developed = pd.DataFrame({'rcid': new_UN_data_df.rcid,
                                  'unres': new_UN_data_df.unres,
                                  'date': new_UN_data_df.date,
                                  'year': new_UN_data_df.year, 
                                  'me': new_UN_data_df.me,
                                  'nu': new_UN_data_df.nu,
                                  'di': new_UN_data_df.di,
                                  'co': new_UN_data_df.co,
                                  'hr': new_UN_data_df.hr,
                                  'ec': new_UN_data_df.ec,
                                  'developed_abstain': 0,
                                  'developed_yes': 0, 
                                 'developed_no': 0, 
                                  'developed_absent': 0,
                                  'developed_not_member': 0,
                                  'undeveloped_abstain': 0,
                                  'undeveloped_yes': 0, 
                                 'undeveloped_no': 0, 
                                  'undeveloped_absent': 0,
                                  'undeveloped_not_member': 0,})
```

```python
vote_by_developed.head()
```

```python
vote_by_developed.drop_duplicates(inplace=True)
vote_by_developed.reset_index(drop=True, inplace=True)
```

```python
vote_by_developed.head() 
```

```python
# iterate over the rcid column
  # subset_rcid = make a subset of the new_UN_data_df of that rcid
  # subset_year = make a subset of the new_UN_data_df of that year of the rcid
  # find the median GDP of that year and store it 

  # developed = create a subset of the of subset_rcid when the developed > median
  # yes_developed = create a subset of the of developed when the vote == 1
  # no_developed = create a subset of the of developed when the vote == 3

  # undeveloped = create a subset of the of subset_rcid when the developed > median
  # yes_undeveloped = create a subset of the of developed when the vote == 1
  # no_undeveloped = create a subset of the of developed when the vote == 3

for res in vote_by_developed.rcid:
  subset_rcid = new_UN_data_df[new_UN_data_df['rcid'] == res]
  year = str(subset_rcid['year'].unique()[0])
  median = new_UN_data_df[new_UN_data_df['year'] == int(year)]['GDP per capita (current US$)'].median()

  index = vote_by_developed[vote_by_developed['rcid'] == res].index.values[0] # index of res in new dataframe

  developed = subset_rcid[subset_rcid['GDP per capita (current US$)'] > median]
  vote_by_developed['developed_yes'][index] = len(developed[developed['vote'] == 1])
  vote_by_developed['developed_abstain'][index] = len(developed[developed['vote'] == 2])
  vote_by_developed['developed_no'][index] = len(developed[developed['vote'] == 3])
  vote_by_developed['developed_absent'][index] = len(developed[developed['vote'] == 8])
  vote_by_developed['developed_not_member'][index] = len(developed[developed['vote'] == 9])

  undeveloped = subset_rcid[subset_rcid['GDP per capita (current US$)'] <= median]
  vote_by_developed['undeveloped_yes'][index] = len(undeveloped[undeveloped['vote'] == 1])
  vote_by_developed['undeveloped_abstain'][index] = len(undeveloped[undeveloped['vote'] == 2])
  vote_by_developed['undeveloped_no'][index] = len(undeveloped[undeveloped['vote'] == 3])
  vote_by_developed['undeveloped_absent'][index] = len(undeveloped[undeveloped['vote'] == 8])
  vote_by_developed['undeveloped_not_member'][index] = len(undeveloped[undeveloped['vote'] == 9])


  
```
